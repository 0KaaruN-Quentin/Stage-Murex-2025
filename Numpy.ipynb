```python
def initialisation (l, c):
    res = np.random.randn(l, c)
    res1 = np.ones((l, 1))
    res = np.concatenate((res, res1), axis = 1)
    return res
```

```python
import imageio.v2 as imageio
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
face = imageio.imread("https://upload.wikimedia.org/wikipedia/commons/thumb/b/b6/Felis_catus-cat_on_snow.jpg/800px-Felis_catus-cat_on_snow.jpg")
plt.imshow(face)
plt.show()
```

```python
y, x = face.shape[:-1]
y, x = y//4, x//4
resized = face[y:-y+1, x:-x+1, :]
plt.imshow(resized)
plt.show()
```

```python
resized = rgb2gray(resized)
recoloured = resized.copy()
recoloured[recoloured < 0.2] = 0
recoloured[recoloured > 0.8] = 1
plt.imshow(recoloured, cmap = "gray")
plt.show
```
